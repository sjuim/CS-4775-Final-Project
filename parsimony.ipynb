```python
from Bio import AlignIO
from Bio.Phylo.TreeConstruction import ParsimonyScorer, NNITreeSearcher, ParsimonyTreeConstructor

# Read the alignment from the fasta file
aln = AlignIO.read("congo_nucleoproteins_padded.fasta", "fasta")

# Create a scorer for parsimony
scorer = ParsimonyScorer()

# Create a tree searcher using nearest-neighbor interchange (NNI)
searcher = NNITreeSearcher(scorer)

# Create a tree constructor using the searcher and starting tree
constructor = ParsimonyTreeConstructor(searcher)

# Build the parsimony tree
pars_tree = constructor.build_tree(aln)



```

```python
import matplotlib
import matplotlib.pyplot as plt
from Bio import Phylo

fig = plt.figure(figsize=(20, 12), dpi=300) # create figure & set the size 
matplotlib.rc('font', size=18)              # fontsize of the leaf and node labels 
matplotlib.rc('xtick', labelsize=16)       # fontsize of the tick labels
matplotlib.rc('ytick', labelsize=16)       # fontsize of the tick labels
axes = fig.add_subplot(1, 1, 1)

# drawing the tree
Phylo.draw(pars_tree, axes=axes)
fig.savefig("Parsimony_tree")
```
